```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from sklearn import cluster
from scipy.cluster import hierarchy
```

```python
def plotImage (img , title):
    plt.figure(figsize = (10 , 10))
    plt.imshow(img)
    plt.title (title)
```

```python
def plotDendogram (img):
    linkage = hierarchy.ward(img)
    plt.figure(figsize = (10, 10))
    hierarchy.dendrogram (linkage)
    plt.title("Dendrogram")
```

```python
#read the image
imgBGR = cv2.imread('image/3063.jpg')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

#scale down for faster computation
scale_percent = 25 # percent of original size
width = int(imgRGB.shape[1] * scale_percent / 100)
height = int(imgRGB.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
img = cv2.resize(imgRGB, dim, interpolation = cv2.INTER_AREA)

image = np.copy (img)
```

```python
plotImage(image, 'original')
```

```python
#use median blur to remove salt noise
image1 = cv2.medianBlur(image, 5)
plotImage(image1, 'after median blur')
```

```python
# reshape from 3D RBG to 2D
# pImage means processed image
pImage = image1.reshape((-1, 3))
pImage.shape
```

```python
plotDendogram(pImage)
```

```python
n = 2
c = cluster.AgglomerativeClustering(n_clusters=n)
c.fit(pImage)
labels = c.labels_
centers = np.zeros((n, 3))
for i in range(n):
    centers[i] = np.mean(pImage[labels == i], axis=0)
 
output = centers[labels].reshape(image.shape)
plotImage(output/255, "Agglomerative Clustering with " + str(n) + " clusters")
```

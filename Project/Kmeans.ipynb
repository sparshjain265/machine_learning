```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from sklearn import cluster
```

```python
def plotImage (img , title):
    plt.figure(figsize = (10 , 10))
    plt.imshow(img)
    plt.title (title)
```

```python
#read the image
imgBGR = cv2.imread('image/41029.jpg')
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
image = np.copy (imgRGB)
```

```python
plotImage(image, 'original')
```

```python
#use median blur to remove salt noise
image1 = cv2.medianBlur(image, 5)
plotImage(image1, 'after median blur')
```

```python
# reshape from 3D RBG to 2D
# pImage means processed image
pImage = image1.reshape((-1, 3))
```

```python
for n in [2, 3, 4, 5]:
    c = cluster.KMeans(n_clusters=n)
    c.fit(pImage)
    centers = c.cluster_centers_
    labels = c.labels_
    output = centers[labels].reshape(image.shape)
    plotImage(output/255, "K = " + str(n))
```
